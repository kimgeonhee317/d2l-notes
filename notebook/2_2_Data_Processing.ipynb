## 2.2.1 Reading the Dataset

```python
# Create csv file as path '../data/house_tiny.csv'
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..','data', 'house_tiny.csv')
with open(data_file , 'w') as f:
  f.write('''NumRooms,RoofType,Price
NA,NA,127500
2,NA,106000
4,Slate,178100
NA,NA,140000''')
```

```python
# Load dataset with Pandas
import pandas as pd
data = pd.read_csv(data_file)
data
```

## 2.2.2. Data Preparation

Missing values might be handled by imputation or deletion.

```python
# We can treat NaN as one of the categorical values as imputing techinque.
inputs, targets = data.iloc[:, 0:2], data.iloc[:, 2] # split between input and target
print(inputs)
inputs = pd.get_dummies(inputs, dummy_na = True) # one-hot encoded nan
inputs
```

```python
# For numerical, replace it is to mean value
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

## 2.2.3 Conversion to the Tensor Format

```python
# now that all the entries are numerical, load them into tensor
import torch

X, y = torch.tensor(inputs.values), torch.tensor(targets.values)
X, y
```
